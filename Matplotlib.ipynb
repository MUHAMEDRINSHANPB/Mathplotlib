```python

x = [1, 2, 3, 4,5]
y = [2, 4, 6, 8, 10]


# Create a line chart
plt.figure(figsize = (8, 6)) #Creates a new figure with a size of 8 inches by 6 inches 
plt.plot(x, y, marker = 'o', linestyle = '-')


# Add annotations
for i, (xi, yi) in enumerate(zip(x, y)):
    plt.annotate(f' ({xi}, {yi})', (xi, yi), textcoords = "offset points", xytext = (0, 10), ha = 'center')


# textcoords :Specifies the coordinate system for the annotation text. "offset points" means the text is offset from the point.
# xytext: Specifies the offset for the text in points (10 points above the data point).
# ha :Horizontal alignment of the text.


# Add title and labels
plt. title('Line Chart with Annotations') 
plt. xlabel('X - axis Label')
plt.ylabel('Y - axis Label')


# Display grid
plt.grid(True)


# Show the plot
plt. show()
```

```python

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]


# Create a line chart
plt.figure(figsize = (8, 6))
plt.plot(x, y, marker = 'o', linestyle = '-')


# Add a basic annotation with an arrow
plt.annotate( 'Annotation with arrow',
            xy = (3, 6), 
            xytext = (4, 7), 
            arrowprops = dict(facecolor = 'red', arrowstyle = '->'))


# Add title and labels
plt.title('Basic Annotation with Arrow')
plt.xlabel('X-axis Label') 
plt.ylabel('Y-axis Label')


# Display grid
plt.grid(True)

# Show the plot
plt.show()
```

```python

# Example 2: Multiple Annotations with Different Styles

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10] 


# Create a line chart
plt.figure (figsize = (8, 6))
plt. plot(x, y, marker = 'o', linestyle = '-')


# Add multiple annotations with different styles
annotations = [
              {'xy': (1, 2), 'text': 'Point 1', 'xytext': (2, 3), 'arrowprops': dict(facecolor = 'blue', arrowstyle = '->')},
              {'xy': (3, 6),'text': 'Point 2','xytext': (4, 5), 'arrowprops': dict(facecolor = 'green', arrowstyle = '->')},
              {'xy': (5, 10),'text': 'Point 3', 'xytext': (4, 11), 'arrowprops': dict(facecolor = 'red', arrowstyle = '->')}
]

for ann in annotations:
    plt.annotate(
            ann['text'],
          xy = ann['xy'],
          xytext = ann['xytext'],
          arrowprops = ann['arrowprops']
    )
    
plt.title('Multiple Annotations with Different Styles') 
plt.xlabel('X-axis Label')
plt.ylabel ('Y-axis Label')

plt.grid(True)

plt.show()
```

```python

# set width of bar
barWidth = 0.25
fig      = plt.subplots(figsize = (12, 8))


# set height of bar
IT  = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]


# Set position of bar on X axis
br1 = np.arange(len (IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]


# Make the plot
plt.bar(br1, IT,  color = 'r', width = barWidth, edgecolor = 'grey', label = 'IT')
plt.bar(br2, ECE, color = 'g', width = barWidth, edgecolor = 'grey', label = 'ECE')
plt.bar(br3, CSE, color = 'b', width = barWidth, edgecolor = 'grey', label = 'CSE')


# Adding Xticks
plt.xlabel ('BRANCH', fontweight = 'bold', fontsize = 15)
plt.ylabel ('Students Passed', fontweight = 'bold', fontsize = 15)
plt.xticks([r + barWidth for r in range (len (IT))], [' 2015', '2016', '2017', '2018', '2019'])


plt.legend()
plt.show()
```

```python

cars = ["AUDI", "BMW", "FORD", "TESLA", "JAGUAR", "MERCEDES"]

data = [23, 17, 35, 29, 12, 41]

explode = (0.1, 0, 0.2, 0.3, 0, 0)
colors = ["orange", "cyan", "brown", "grey", "indigo", "beige"]

plt.pie(data, explode = explode, labels = cars, colors = colors,
        autopct = "%.1f%%", shadow = True, startangle = 90)

plt.legend(cars, title = "cars", loc = "center left", 
           bbox_to_anchor = (1, 0, 0.5, 1))

plt.title("Customizing Pie Chart")

plt.show()
```
